````
import pandas as pd 



url = 'https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/stocks/Travel.xlsx'
xls = pd.ExcelFile(url)
# df1 = pd.read_excel(xls, sheet_name='008770')
# sheet name 008770ÏùÄ Ïù¥Î¶ÑÍ≤ÄÏÉâÏù¥ ÏïàÎêúÎã§Í≥† ÎÇòÏò¥ 
df1= pd.read_excel(xls,sheet_name=0)
df2 = pd.read_excel(xls,sheet_name=1)
df1
````

````
df2
````

````
df1 = df1.transpose()
new_header = df1.iloc[0]
df1 = df1.iloc[1:]
df1.columns = new_header
df1
````

````
df2 = df2.transpose()
new_header = df2.iloc[0]
df2 = df2.iloc[1:]
df2.columns = new_header
df2
````
